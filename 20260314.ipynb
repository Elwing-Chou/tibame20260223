````python
import urllib.request as req
import json
import pandas as pd

url = "https://api.gamer.com.tw/anime/v1/danmu.php?videoSn=35241&geo=TW%2CHK"
f = req.urlopen(url)
content = f.read()
# print(type(content))
content = json.loads(content)
# print(type(content))
danmu = content["data"]["danmu"]
# for d in danmu:
#     print(d["userid"], d["text"])

# pandas自訂型態: 1. DataFrame: 表格  2. ?
df = pd.DataFrame(danmu)
df.to_csv("danmu.csv", encoding="utf-8")
df.to_excel("danmu.xlsx")
````

````python
import urllib.request as req
import json
import pandas as pd

def download_danmu(serial, fp):
    pat = "https://api.gamer.com.tw/anime/v1/danmu.php?videoSn={}&geo=TW%2CHK"
    url = pat.format(serial)
    print(url)
    f = req.urlopen(url)
    content = f.read()
    content = json.loads(content)
    danmu = content["data"]["danmu"]

    df = pd.DataFrame(danmu)
    df.to_csv(fp, encoding="utf-8")

download_danmu(35242, "xzxxxxx.csv")
````
