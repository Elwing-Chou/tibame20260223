```python
import urllib.request as req
import json
import pandas as pd

# name: 1 2 3...
def download_danmu(serial, fp, episode):
    pat = "https://api.gamer.com.tw/anime/v1/danmu.php?videoSn={}&geo=TW%2CHK"
    url = pat.format(serial)
    print(url)
    f = req.urlopen(url)
    content = f.read()
    content = json.loads(content)
    danmu = content["data"]["danmu"]

    df = pd.DataFrame(danmu)
    # 增加一行, 並且設定成同一個值 取行: df[行名稱]
    df["episode"] = episode
    df.to_csv(fp, encoding="utf-8")

download_danmu(35242, "xzxxxxx.csv", "10")
```

```python
import os

import urllib.request as req
import bs4 as bs

h = {
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/145.0.0.0 Safari/537.36 Edg/145.0.0.0",
}

# url = "https://ani.gamer.com.tw/animeVideo.php?sn=35241"
url = "https://github.com/Elwing-Chou/tibame20260223/raw/refs/heads/main/baha.html"
r = req.Request(url, headers=h)
resp = req.urlopen(r)
content = resp.read()
html = bs.BeautifulSoup(content)

save_dir = "baha"
# 新創資料夾: 如果資料夾不存在, make起來
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

for a in html.find_all("a"):
    # 找到對象區塊以後: 屬性(["href"]) 文字(.get_text())
    href = a["href"]
    if href.startswith("?sn="):
        print("下載:", href)
        sn = href.replace("?sn=", "")
        episode = a.get_text()
        print("集數:", episode)
        fp = save_dir + "/" + sn + ".csv"
        print("儲存位置:", fp)
        download_danmu(sn, fp, episode)
```

```python
# 把所有csv聚集成一個大的表格
# 針對每一個出現過的userid
# 把他發過的danmu拿出來

import glob

total_df = []
for fp in glob.glob("baha/*.csv"):
    df = pd.read_csv(fp, encoding="utf-8", index_col=0)
    total_df.append(df)

total = pd.concat(total_df, axis=0)
total = total.sort_values("episode")
# 取行: [行名稱]
# 取獨特: unique()/value_counts()
users = total["userid"].unique()

# drive/MyDrive/result
save_dir = "result"
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

for u in users:
    fil = total["userid"] == u
    fp = save_dir + "/" + u + ".csv"
    total[fil].to_csv(fp, encoding="utf-8", index=False)
```
