````python
import urllib.request as req
import bs4 as bs

def get_tabelog_page(num):
    # 1. 先準備list
    table = []

    url = "https://tabelog.com/tw/tokyo/rstLst/sweets/" + str(num) + "/?SrtT=rt"
    print(url)
    resp = req.urlopen(url)
    content = resp.read()
    html = bs.BeautifulSoup(content)

    # find/find_all
    rs = html.find_all("div", {"class":"list-rst__body"})
    for r in rs:
        # 拿到
        # 地點/type 評價 晚間價格 午間價格
        en = r.find("a", {"class":"list-rst__rst-name-target"})
        area_genre = r.find("div", {"class":"list-rst__area-genre"})
        rating = r.find("span", {"class":"c-rating__val"})
        prices = r.find_all("span", {"class":"c-rating-v3__val"})
        lunch = prices[0]
        dinner = prices[1]

        # 屬性/顯示: ["href"]/.get_text()
        en_href = en["href"]
        en_text = en.get_text().strip()
        area_genre_text = area_genre.get_text().strip()
        rating_text = rating.get_text().strip()
        lunch_text = lunch.get_text().strip()
        dinner_text = dinner.get_text().strip()

        # 印(ctrl + / 多行註解)
        # print(rating_text)
        # print(en_text)
        # print(en_href)
        # print(area_genre_text)
        # print(lunch_text)
        # print(dinner_text)

        # 2. 準備每一筆資料
        row = {
            "rating":rating_text,
            "name":en_text,
            "url":en_href,
            "area_genre":area_genre_text,
            "lunch_price":lunch_text,
            "dinner_price":dinner_text
        }
        table.append(row)
    return table
````

````python
# 只要做存檔, 就把格式變成JSON格式
# [danmu, danmu, danmu]  danmu: {}
import pandas as pd
total = []
for i in range(10):
    page = i + 1
    total = total + get_tabelog_page(page)

df = pd.DataFrame(total)
df.to_csv("tabelog.csv", encoding="utf-8")
````
