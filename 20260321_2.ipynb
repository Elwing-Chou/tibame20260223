````python
import urllib.request as req
import bs4 as bs

url = "https://tabelog.com/tw/tokyo/rstLst/sweets/?SrtT=rt"
resp = req.urlopen(url)
content = resp.read()
html = bs.BeautifulSoup(content)
print(html)
````
