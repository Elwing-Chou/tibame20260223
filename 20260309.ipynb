```python
# 創造一個清單
# 1. 同樣元素
l = [0] * 10
# 2. 不同元素
# 標準做法
l = []
for i in range(10):
    l.append(i+1)
# 簡便作法
l = [i+1 for i in range(10)]
l
```

```python
# 創造二維清單
# 這個做法不要
a = [0] * 10
l = [a] * 3
l[0][0] = 999
print(l)

# 正確作法1
l = []
for i in range(3):
    a = [0] * 10
    l.append(a)

l[0][0] = 888
print(l)

# 簡便做法
l = [[0] * 10 for i in range(3)]
l[0][0] = 777
print(l)
```

```python
# [3, 4, 5, 6, 7]
# [12, 11, 10, 9, 8],
# [13, 14, 15, 16, 17]
# [22, 21, 20, 19, 18]
start = 3
row, col = 8, 7
l = [[0] * col for i in range(row)]

for i in range(row):
    for j in range(col):
        if i % 2 == 0:
            l[i][j] = col * i + j + start
        else:
            # col * i + 4 ->9,19,29
            l[i][j] = (col * i + col - 1) - j + start
l
```
