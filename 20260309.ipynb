```python
# 創造一個清單
# 1. 同樣元素
l = [0] * 10
# 2. 不同元素
# 標準做法
l = []
for i in range(10):
    l.append(i+1)
# 簡便作法
l = [i+1 for i in range(10)]
l
```

```python
# 創造二維清單
# 這個做法不要
a = [0] * 10
l = [a] * 3
l[0][0] = 999
print(l)

# 正確作法1
l = []
for i in range(3):
    a = [0] * 10
    l.append(a)

l[0][0] = 888
print(l)

# 簡便做法
l = [[0] * 10 for i in range(3)]
l[0][0] = 777
print(l)
```

```python
# [3, 4, 5, 6, 7]
# [12, 11, 10, 9, 8],
# [13, 14, 15, 16, 17]
# [22, 21, 20, 19, 18]
start = 3
row, col = 8, 7
l = [[0] * col for i in range(row)]

for i in range(row):
    for j in range(col):
        if i % 2 == 0:
            l[i][j] = col * i + j + start
        else:
            # col * i + 4 ->9,19,29
            l[i][j] = (col * i + col - 1) - j + start
l
```

```python
# 群集型態:
# 清單(list): [60, 80, 40] 用途: 放置多個同類型東西
# 字典(dict): {"name":"Elwing", "height":175} 用途: 聚集多個欄位來表示一個複雜東西
# 集合(set): {}
# 元組(tuple): ("Elwing", 175) 用途: 簡化版本的字典 key=0 1 2
color = {"r":255, "g":128, "b":200}
color = (255, 128, 200)
print(color)

# 型態會被分成兩種型態: 1. 可變型態(可延伸/可設置):list/dict 2. 不可變型態: tuple
l = [20, 30, 40]
l[0] = 99

# !!!! 要能當作key的東西必須是個不可變型態
# tuple: 當作key來使用
dic = {
    (0, 0):20
}
print(dic)
```
