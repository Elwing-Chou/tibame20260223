```python
# 創造一個清單
# 1. 同樣元素
l = [0] * 10
# 2. 不同元素
# 標準做法
l = []
for i in range(10):
    l.append(i+1)
# 簡便作法
l = [i+1 for i in range(10)]
l
```

```python
# 創造二維清單
# 這個做法不要
a = [0] * 10
l = [a] * 3
l[0][0] = 999
print(l)

# 正確作法1
l = []
for i in range(3):
    a = [0] * 10
    l.append(a)

l[0][0] = 888
print(l)

# 簡便做法
l = [[0] * 10 for i in range(3)]
l[0][0] = 777
print(l)
```
