```python
# 想法2
def is_leap_year(year):
    # TAB/shift+TAB
    # year = 2000
    if year % 400 == 0:
        return True
    elif year % 100 == 0:
        return False
    elif year % 4 == 0:
        return True
    else:
        return False

def is_date_valid(d):
    # d = 20240229
    # 243 % 100 = 43
    # 243 // 100 = 2
    day = d % 100
    remain = d // 100
    month = remain % 100
    year = remain // 100
    # print(year, month, day)

    # 年檢定/月檢定
    # 十二個月該怎麼剪定日
    if year >= 1:
        if 1 <= month <= 12:
            days_per_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
            day_max = days_per_month[month-1]
            # if month == 2 and is_leap_year(year) == True
            if month == 2 and is_leap_year(year):
                day_max = day_max + 1
            if 1 <= day <= day_max:
                return True
    return False

def count_fully_squared_2(start=4499, maxcount=10):
    # 20241001 = 4499 ** 2
    count = 0
    # start = 4499

    result = []
    while True:
        d = start ** 2
        if is_date_valid(d):
            # print(d)
            result.append(d)
            count = count + 1
        if count == maxcount:
            break
        start = start + 1
    return result
```

```python
# 想法1
import datetime as dt

def is_sqrt_int(n):
    # n = 104976
    n_sqrt = n ** 0.5
    n_sqrt_round = round(n_sqrt)
    if abs(n_sqrt-n_sqrt_round) <= 10 ** -6:
        return True
    else:
        return False

def count_fully_squared_1(year=2024, month=10, day=1, maxcount=10):
    # 開始日期
    start = dt.datetime(year, month, day)
    delta = dt.timedelta(days=1)
    # 往後迴圈
    count = 0
    result = []
    while True:
        # 日期->字串(20241001)->整數
        n = int(start.strftime("%Y%m%d"))
        # 你要確認整數開根號是否是整數(下面說)
        if is_sqrt_int(n):
            # 如果是整數就是個完全平方日
            # print(n)
            result.append(n)
            count = count + 1
        if count == maxcount:
            break
        start = start + delta
    return result

count_fully_squared_1()
```

```python
import time

count = 100

t1 = time.time()
count_fully_squared_1(maxcount=count)
t2 = time.time()
inter1 = t2 - t1
print("執行多久:", inter1)

t1 = time.time()
count_fully_squared_2(maxcount=count)
t2 = time.time()
inter2 = t2 - t1
print("執行多久:", inter2)

print(inter1 / inter2)
```

```python
# 象棋棋盤: https://i-blog.csdnimg.cn/blog_migrate/ee144a254459ba034fae2ff541b13d64.png
import pygame as pg

#pygame初始化
pg.init()

# --- 字型 ---
def get_font(size):
    for f in ['microsoftjhenghei', 'simhei', 'stheitirelight']:
        if f in pg.font.get_fonts():
            return pg.font.SysFont(f, size)
    return pg.font.SysFont(None, size)

# 32是字體的size
FONT_UI = get_font(32)

#設定視窗
width, height = 800, 880
inter = 80
# 產生視窗
screen = pg.display.set_mode([width, height])
# 設定遊戲標題
pg.display.set_caption("象棋")

# 建立畫布bg
bg = pg.Surface(screen.get_size())
# 把畫布填滿某個顏色
bg.fill([199, 167, 82])

# 畫上半部棋盤
# line(畫布, 顏色, 開始, 結束, 線寬度)
for i in range(5):
    pg.draw.line(bg,
                 [0, 0, 0],
                 [inter, inter*i+inter],
                 [9*inter, inter*i+inter],
                 2)

for i in range(9):
    pg.draw.line(bg,
                 [0, 0, 0],
                 [inter*i+inter, inter],
                 [inter*i+inter, 5*inter],
                 2)

pg.draw.line(bg,
             [0, 0, 0],
             [4*inter, inter],
             [6*inter, 3*inter],
             2)

# 你要把圖層放到上一層
screen.blit(bg, [0, 0])
# 對畫面進行更新(才會真的秀出來)
pg.display.update()

# 建立一個永不結束的迴圈(遊戲才不會結束)
running = True
while running:
    # 收取你的遊戲任何事件(滑鼠點擊/鍵盤按鈕...)
    for event in pg.event.get():
        # 偵測滑鼠點擊以後放掉的動作
        if event.type == pg.MOUSEBUTTONUP:
            pass
        # 如果收到的事件是按x
        if event.type == pg.QUIT:
            # 迴圈就會變成while False
            running = False

pg.quit()
```
