```python
# 想法2
def is_leap_year(year):
    if year % 400 == 0:
        return True
    elif year % 100 == 0:
        return False
    elif year % 4 == 0:
        return True
    else:
        return False

def is_date_valid(d):
    day = d % 100
    remain = d // 100
    month = remain % 100
    year = remain // 100
    # print(year, month, day)
    # 確認是否合理
    if year >= 1:
        if 1 <= month <= 12:
            day_per_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
            day_max = day_per_month[month-1]
            # if month == 2 and is_leap_year(year) == True
            if month == 2 and is_leap_year(year):
                day_max = day_max + 1
            if 1 <= day <= day_max:
                return True
    return False

def get_fully_squared_2(start=4499, limit=10):
    count = 0
    result = []
    while True:
        d = start ** 2
        if is_date_valid(d):
            result.append(d)
            count = count + 1
        if count == limit:
            break
        start = start + 1
    return result
```

```python
# 變成一個函是
import datetime as dt
def get_fully_squared_1(y=2024, m=10, d=1, limit=10):
    d = dt.datetime(y, m, d)
    delta = dt.timedelta(days=1)
    # 寫一個迴圈 把d開根號 確認他是否是整數
    # 往下數十個
    result = []
    count = 0
    while True:
        d_str = d.strftime("%Y%m%d")
        d_int = int(d_str)
        d_int_root = d_int ** 0.5
        if d_int_root.is_integer():
            result.append(d_int)
            count = count + 1
        if count == limit:
            break
        d = d + delta
    return result

get_fully_squared_1()
```

```python
import time

l = 200

t1 = time.time()
get_fully_squared_1(limit=l)
t2 = time.time()
inter1 = t2 - t1

t1 = time.time()
get_fully_squared_2(limit=l)
t2 = time.time()
inter2 = t2 - t1

print("想法1:", inter1)
print("想法2:", inter2)
print("差多少倍:", inter1 / inter2)
```

```python
# 象棋棋盤: https://i-blog.csdnimg.cn/blog_migrate/ee144a254459ba034fae2ff541b13d64.png
import pygame as pg

#pygame初始化
pg.init()

# --- 字型 ---
def get_font(size):
    for f in ['microsoftjhenghei', 'simhei', 'stheitirelight']:
        if f in pg.font.get_fonts():
            return pg.font.SysFont(f, size)
    return pg.font.SysFont(None, size)

# 32是字體的size
FONT_UI = get_font(32)

#設定視窗
width, height = 800, 880
inter = 80
# 產生視窗
screen = pg.display.set_mode([width, height])
# 設定遊戲標題
pg.display.set_caption("象棋")

# 建立畫布bg
bg = pg.Surface(screen.get_size())
# 把畫布填滿某個顏色
bg.fill([199, 167, 82])


# 你要把圖層放到上一層
screen.blit(bg, [0, 0])
# 對畫面進行更新(才會真的秀出來)
pg.display.update()

# 建立一個永不結束的迴圈(遊戲才不會結束)
running = True
while running:
    # 收取你的遊戲任何事件(滑鼠點擊/鍵盤按鈕...)
    for event in pg.event.get():
        # 偵測滑鼠點擊以後放掉的動作
        if event.type == pg.MOUSEBUTTONUP:
            pass
        # 如果收到的事件是按x
        if event.type == pg.QUIT:
            # 迴圈就會變成while False
            running = False

pg.quit()
```
